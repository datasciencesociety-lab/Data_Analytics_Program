This is where Exploratory Data Analysis (EDA) begins.

EDA is the process of exploring, understanding, and analyzing data to uncover meaningful patterns that help answer business questions and support decision-making.

---

# Why Do Businesses Use Data?

Every organization generates data through its daily operations.

This data captures information about customers, employees, sales, products, finances, and many other aspects of the business.

Simply collecting data does not create value.

The real value comes from analyzing that data to answer important business questions such as:

- Why are sales declining?
- Which products perform the best?
- Why are customers leaving?
- Which marketing campaign generated the highest revenue?
- Which employees are most likely to resign?

The answers to these questions help organizations make informed decisions rather than relying on assumptions.

---

# The Analytics Journey

Every data analytics project typically follows a logical sequence.

```text
Business Problem
        ↓
Collect Data
        ↓
Prepare Data
        ↓
Exploratory Data Analysis (EDA)
        ↓
Insights
        ↓
Business Decision
```

Let's briefly understand each stage.

### 1. Business Problem

Every project starts with a business challenge.

Examples include:

- Employee attrition is increasing.
- Customer satisfaction has dropped.
- Sales have declined.
- Delivery times are becoming longer.

Without a business problem, there is no analysis to perform.

---

### 2. Collect Data

Once the problem is identified, relevant data is collected.

The data may come from:

- Company databases
- Surveys
- Websites
- APIs
- Excel files
- CSV files

The quality of analysis depends heavily on the quality of the collected data.

---

### 3. Prepare Data

Raw data is rarely ready for analysis.

Before exploring it, we often need to:

- Handle missing values
- Remove duplicate records
- Correct inconsistent values
- Identify outliers
- Convert data into appropriate formats

A well-prepared dataset makes the analysis more reliable.

---

### 4. Exploratory Data Analysis (EDA)

After the data is prepared, we begin exploring it.

During EDA, we try to answer business questions by:

- Studying distributions
- Comparing different groups
- Finding relationships
- Identifying trends
- Detecting unusual patterns

Visualizations and statistical techniques help us understand the data, but they are tools—not the final goal.

---

### 5. Insights

An insight is a meaningful conclusion supported by data.

For example:

Instead of saying,

> "The Sales department has the highest attrition."

a better insight would be,

> "The Sales department experiences significantly higher attrition than other departments, suggesting possible workload or job satisfaction issues."

Insights explain **what the data is telling us**.

---

### 6. Business Decision

The purpose of analysis is to support better decision-making.

Examples include:

- Increase employee engagement programs.
- Revise overtime policies.
- Improve customer support.
- Launch targeted marketing campaigns.
- Allocate resources more efficiently.

This is where the business gains value from data.

---

# Where Does EDA Fit?

Many beginners believe that creating graphs is the primary objective of EDA.

In reality, EDA is only one step in the overall analytics process.

Its purpose is to understand the data well enough to generate meaningful insights that support business decisions.

EDA acts as the bridge between prepared data and actionable recommendations.

---

# Key Takeaways

- Businesses analyze data to solve real-world problems.
- Every analytics project begins with a business objective.
- Data must be prepared before meaningful analysis can begin.
- EDA helps us understand the data and answer business questions.
- Insights generated through EDA support informed business decisions.
- The ultimate goal of analytics is not creating visualizations—it is enabling better business decisions.

# Thinking Like a Data Analyst

A data analyst is not simply someone who knows Python, SQL, or visualization libraries.

A data analyst is someone who investigates business problems using data.

Instead of asking,

> "Which chart should I create?"

an analyst asks,

> "What business question am I trying to answer?"

This small change in thinking makes a huge difference.

The analysis should always be driven by the business problem, not by the available tools.

---

# An Investigator's Mindset

Think about how a detective solves a case.

A detective does not randomly collect evidence or question everyone.

Instead, they:

- Understand the case.
- Observe the available evidence.
- Ask relevant questions.
- Look for patterns.
- Reach conclusions based on evidence.

A data analyst follows a very similar approach.

Rather than investigating a crime, they investigate data.

The goal is to collect evidence from the data that helps answer business questions and support better decisions.

---

# Observation vs Insight

One of the most important skills in EDA is understanding the difference between an observation and an insight.

### Observation

An observation is something we notice in the data.

Example:

> The Sales department has the highest employee attrition.

This is simply a finding from the analysis.

---

### Insight

An insight explains why the observation matters to the business.

Example:

> The Sales department experiences noticeably higher attrition than other departments, suggesting that retention efforts should be prioritized for this department.

Insights help businesses make informed decisions.

---

# Asking the Right Questions

Every analysis begins with questions.

For example, consider a company experiencing high employee attrition.

Some useful questions might be:

- Which department has the highest attrition?
- Does overtime influence attrition?
- Does salary affect employee retention?
- Does work-life balance impact attrition?
- Are experienced employees leaving more frequently?

Notice that these are **business questions**, not coding questions.

Once the questions are clear, choosing the appropriate analysis becomes much easier.

---

# Types of Analysis

After identifying a business question, the next step is to decide **how the data should be analyzed**.

In Exploratory Data Analysis, we commonly perform three types of analysis.

---

## 1. Univariate Analysis

Univariate analysis focuses on **one variable at a time**.

Its objective is to understand the characteristics of a single feature.

Examples:

- Distribution of employee age
- Distribution of monthly income
- Number of employees in each department

---

## 2. Bivariate Analysis

Bivariate analysis examines the relationship between **two variables**.

Its objective is to understand whether one variable is associated with another.

Examples:

- Overtime vs Attrition
- Salary vs Attrition
- Department vs Attrition

---

## 3. Multivariate Analysis

Multivariate analysis studies **three or more variables together**.

Its objective is to understand more complex relationships that cannot be explained by only one or two variables.

Examples:

- Overtime, Department and Attrition
- Salary, Experience and Attrition
- Age, Income and Job Satisfaction

---

# An Important Note

Many courses teach EDA as:

```text
Univariate
      ↓
Bivariate
      ↓
Multivariate
```

This often creates the impression that every EDA project must follow this sequence.

In practice, this is **not** how data analysis works.

The type of analysis depends entirely on the business question.

Sometimes you may begin with a bivariate analysis.

Sometimes a multivariate analysis may be required immediately.

You may also switch between univariate, bivariate, and multivariate analysis multiple times during the same project.

There is **no fixed order**.

The business problem determines the analysis—not the other way around.

---

# Key Takeaways

- A data analyst investigates business problems using data.
- Good analysis begins with meaningful business questions.
- An observation describes what happened, while an insight explains why it matters.
- Univariate, bivariate, and multivariate analysis are different ways of exploring data.
- These are **types of analysis**, not mandatory steps of EDA.
- The business question determines which type of analysis should be used.

````python
attrition_overtime = pd.crosstab(df["OverTime"],df["Attrition"])
````

````python
attrition_overtime
````

````python
sns.countplot(data = df,x = "OverTime",hue = "Attrition")
````

# Null hypothesis: There is no relationship between the overtime and attrition
# ALternate hypothesis : yes there is a relation between the overtime and attrition

````python
def hypothesis_test(p_value):
    alpha = 0.05

    if p_value < alpha:
        print("Reject the Null Hypothesis")
        print("There is a significant relationship between the variables.")
    else:
        print("Fail to Reject the Null Hypothesis")
        print("There is no significant relationship between the variables.")
````

````python
from scipy.stats import chi2_contingency
chi2_stat, p_val, dof, expected = chi2_contingency(attrition_overtime)

print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"p-value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")

hypothesis_test(p_val)
````

# H0 there is no signifact relationship between the monthly income and attrition
# H1 there is a significant realtion between the monthly income and attrition lower salaried employee are more prone to leave the company than the higher salaried employee

````python
from scipy.stats import ttest_ind

yes_income = df[df["Attrition"] == "Yes"]["MonthlyIncome"]
no_income = df[df["Attrition"] == "No"]["MonthlyIncome"]

t_stat, p_value = ttest_ind(yes_income, no_income)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
hypothesis_test(p_value)
````
