# 1. Reshaping Arrays

## Definition

Reshaping is the process of changing the dimensions of an array without changing its data.

---

## Working

Sometimes arrays are stored in one structure but needed in another structure.

Example:

```python
[1, 2, 3, 4, 5, 6]
```

can be reshaped into:

```python
[
 [1, 2, 3],
 [4, 5, 6]
]
```

The total number of elements must remain the same.

---

## Syntax

```python
array.reshape(rows, columns)
```

---

## Example

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

reshaped = arr.reshape(2, 3)

print(reshaped)
```

## Output

```python
[[1 2 3]
 [4 5 6]]
```

---

# 2. Flattening Arrays

## Definition

Flattening converts a multi-dimensional array into a one-dimensional array.

---

## Working

A 2D array:

```python
[
 [1, 2],
 [3, 4]
]
```

becomes:

```python
[1, 2, 3, 4]
```

Flattening helps simplify arrays into a single row of elements.

---

## Syntax

```python
array.flatten()
```

---

## Example

```python
import numpy as np

matrix = np.array([
    [1, 2],
    [3, 4]
])

print(matrix.flatten())
```

## Output

```python
[1 2 3 4]
```

---

# 3. Combining Arrays

## Definition

Combining arrays means joining multiple arrays together into a single array.

---

## Working

NumPy allows arrays to be joined:
- vertically
- horizontally

Example:

```python
[1, 2]
```

and

```python
[3, 4]
```

can become:

```python
[1, 2, 3, 4]
```

---

## Syntax

```python
np.concatenate((array1, array2))
```

---

## Example

```python
import numpy as np

arr1 = np.array([1, 2])
arr2 = np.array([3, 4])

combined = np.concatenate((arr1, arr2))

print(combined)
```

## Output

```python
[1 2 3 4]
```

# 3. Using Seed Values

## Definition

A seed value fixes the random output so the same random numbers are generated every time.

---

## Working

Normally random numbers change in every execution.

Using:

```python
np.random.seed()
```

makes the random output repeatable.

This is useful when:
- testing programs
- debugging
- sharing results

---

## Syntax

```python
np.random.seed(value)
```

---

## Example

```python
import numpy as np

np.random.seed(1)

print(np.random.randint(1, 10, 5))
```

## Output

```python
[6 9 6 1 1]
```

The output remains the same every time this code runs.

# Syntax

df.iloc[row_selection, column_selection]

This means:

* Rows from position `0` to `4`

# Filtering & Boolean Indexing using Iris Dataset

In real-world datasets, we usually do not need every row.

For example:

* Flowers with petal length greater than 5
* Flowers belonging to a particular species
* Flowers with large sepal width

To get only useful rows, we use filtering.

---

# What is Filtering?

Filtering means selecting rows that satisfy a condition.

Simple meaning:

> Show only the rows that match the requirement.

---

# Boolean Values

Filtering works using:

* `True`
* `False`

If a condition becomes `True`, that row is selected.

If it becomes `False`, that row is ignored.

---

# Filtering with Single Condition

## Example

```python id="yhnc20"
iris[iris["petal_length"] > 5]
```

This means:

> Show flowers where petal length is greater than 5.

---

# Another Example

```python id="ql7z3o"
iris[iris["species"] == "setosa"]
```

This means:

> Show only Setosa flowers.

---

# Multiple Conditions

Sometimes one condition is not enough.

We may need:

* Petal length greater than 5 AND species is virginica
* Sepal width greater than 3 OR petal width greater than 2

For this, Pandas provides:

* `&` → AND
* `|` → OR

---

# AND Condition

Both conditions must be true.

## Example

```python id="rqyo1m"
iris[(iris["petal_length"] > 5) & (iris["species"] == "virginica")]
```

This selects:

* Flowers with petal length greater than 5
* AND species should be virginica

---

# OR Condition

At least one condition must be true.

## Example

```python id="jlwm44"
iris[(iris["sepal_width"] > 3) | (iris["petal_width"] > 2)]
```

This selects flowers where:

* Sepal width is greater than 3
  OR
* Petal width is greater than 2

---

# Important Rule

Always use parentheses around conditions.

Correct:

```python id="7e5tkt"
iris[(iris["petal_length"] > 5) & (iris["species"] == "virginica")]
```

Wrong:

```python id="2rj5yw"
iris[iris["petal_length"] > 5 & iris["species"] == "virginica"]
```

---

# Common Comparison Operators

| Operator | Meaning                  |
| -------- | ------------------------ |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |

---

# Simple Understanding

Filtering is like saying:

> "Show only flowers with large petals."

or

> "Show only Setosa flowers."

# Why Do We Perform Data Cleaning?

Imagine you are given a dataset from a company.

Your manager asks:

* What is the average employee salary?
* Which city has the most employees?
* Which department performs best?

Can we directly start analyzing the data?

The answer is usually **No**.

Real-world data is rarely perfect. Before analysis, we must understand the data and identify potential issues that may affect our results.

This process is called **Data Cleaning**.

---

# What Type of Data Problems Do We Usually Face?

In real-world projects, analysts mainly encounter two types of problematic data:

## 1. Dirty Data

Dirty data refers to data that contains errors, incorrect values, or poor-quality information.

Examples:

* Missing values
* Duplicate records
* Incorrect data types
* Invalid values
* Inconsistent text

Example:

| Name  | Age | City   |
| ----- | --- | ------ |
| John  | 25  | Delhi  |
| Sarah | NaN | Mumbai |
| John  | 25  | Delhi  |

Problems:

* Missing value in Age
* Duplicate record for John

This is an example of **dirty data** because the data contains quality issues.

---

## 2. Messy Data

Messy data refers to data that is poorly organized, inconsistent, or difficult to work with.

Examples:

* Extra spaces
* Different text formats
* Inconsistent naming conventions
* Columns stored in the wrong format

Example:

| City   |
| ------ |
| Mumbai |
| MUMBAI |
| mumbai |
| Mumbai |

Although all values represent the same city, they are written differently.

This is an example of **messy data** because the information is not standardized.

---

# Dirty Data vs Messy Data

Dirty Data focuses on:

* Missing values
* Duplicate records
* Incorrect values
* Wrong data types

Messy Data focuses on:

* Inconsistent formatting
* Poor organization
* Extra spaces
* Different text representations

Both types of problems can lead to incorrect analysis and misleading results.

Therefore, before performing analysis, we must identify and fix these issues.

---

# Data Cleaning is Not Just Applying Functions

Many beginners think data cleaning means:

```python
dropna()
fillna()
drop_duplicates()
```

and the dataset becomes clean.

This is not true.

The most important part of data cleaning is **making decisions**.

Functions are only tools.

The real skill is deciding when and how to use those tools.

Before applying any cleaning technique, we should understand:

* Why is the data missing?
* Is the missing data important?
* Is a duplicate row actually a mistake?
* Should we remove it or keep it?
* Is a data type incorrect?

# 1. Handling Missing Values

## Why Do We Need It?

Missing values are one of the most common data quality issues.

Examples:

- Age is missing
- Salary is missing
- Department is missing

Missing values can:

- Affect calculations
- Produce incorrect statistics
- Cause errors during analysis
- Reduce data quality

Therefore, we need functions to identify and handle missing data.

---

## Functions Used

### isnull()

Checks whether a value is missing.

Returns:

- True → Missing Value
- False → Not Missing

---

### notnull()

Checks whether a value is present.

Returns:

- True → Value Exists
- False → Missing Value

---

### dropna()

Removes rows or columns containing missing values.

Use when:

- Missing records are not useful
- Very few rows contain missing values

---

### fillna()

Replaces missing values with another value.

Use when:

- We do not want to lose data
- Missing values can be reasonably estimated

Example:

- Replace missing age with average age
- Replace missing city with "Unknown"

---

# 2. Handling Duplicate Records

## Why Do We Need It?

Sometimes the same record appears multiple times.

Example:

Duplicate records can:

- Affect averages

Therefore, we need functions to identify and remove duplicates.

---

## Functions Used

### duplicated()

Checks whether a row is duplicated.

Returns:

- True → Duplicate Row

---

### drop_duplicates()

Removes duplicate rows from the dataset.

Use when:

- The repeated record provides no additional information

# 4. String Cleaning

## Why Do We Need It?

Text data is often inconsistent.

Example:

```text

 MUMBAI

```

All represent the same city but are written differently.

Such inconsistencies can:

- Create duplicate categories
- Affect grouping operations
- Produce incorrect reports

Therefore, we need string cleaning functions.

---

# Why Do We Use .str ?

When working with a DataFrame column, Pandas stores text data as a Series.

A Series contains many values.

Example:

```python
df["City"]
```

contains multiple city names.

We cannot directly write:

```python
df["City"].lower()
```

because `.lower()` is a Python string method that works on a single string, not an entire Pandas column.

Instead, Pandas provides the `.str` accessor.

The `.str` accessor allows us to apply string operations to every value in a column.

Example:

```python
df["City"].str.lower()
```

Meaning:

- Access the City column
- Treat it as text data
- Apply `lower()` to every value

Think of `.str` as:

> "I want to perform string operations on this entire column."

---

## Functions Used

### str.lower()

Converts all text to lowercase.

Example:

```text
MUMBAI
```

↓

```text
mumbai
```

---

### str.upper()

Converts all text to uppercase.

Example:

```text
mumbai
```

↓

```text
MUMBAI
```

---

### str.strip()

Removes spaces from the beginning and end of text.

Example:

```text
" Mumbai "
```

↓

```text
"Mumbai"
```

---

### str.replace()

Replaces one text value with another.

Example:

```text
Hyd
```

↓

```text
Hyderabad
```

Use when:

- Values are inconsistent
- Abbreviations need standardization
- Spelling needs correction
