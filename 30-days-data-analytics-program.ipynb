To work with data in any programming language, we need to store that data in the computer’s memory. However, we cannot access memory directly. So, we assign a name to a memory location where the data is stored. This allows us to easily access and use the data whenever needed, as long as it remains in memory.

This named memory location is called a **variable**.

**Definition:**
A variable is a named memory location used to store data.


````python
a = 10
````

`a = 10`

In the above code, **`a`** is the variable name, and **`10`** is the value assigned to it. This means the value 10 is stored in a memory location, and we can access it later using the variable name `a`.


````python
print(a)
````

To display a value on the screen in Python, we use the `print()` function. By passing the variable name inside the `print()` function, we can see the value stored in that variable.

For example, if we write `print(a)`, it will display the value assigned to `a` on the screen.


## Data Types

As the name suggests, data types tell us the type of data stored in a variable.

In Python, the most commonly used data types are:

* **int** → used for integer values
  Example: `10`, `25`, `-7`

* **float** → used for decimal values
  Example: `10.5`, `3.14`, `-2.7`

* **str** → used for text or characters
  Example: `"hello"`, `"Python"`, `"A"`

* **bool** → used for boolean values (True or False)
  Example: `True`, `False`

Python does not have a separate **char** data type because a single character is also considered a string.

Similarly, unlike some other programming languages:

* There is no separate **long int** or **short int** — all integer values come under **int**
* There is no separate **double** or **long double** — decimal values come under **float**

This makes Python simpler, cleaner, and easier to learn for beginners.


````python
a = 10
b = 3.14
c = "Hello World!!"
d = True
````

````python
print("Data type of a = ",type(a)) #Int Data type
print("Data type of b = ",type(b)) #Float Data type
print("Data type of c = ",type(c)) #String Data type
print("Data type of d = ",type(d)) #boolean Data Type
````

The `type()` function is used to find the data type of a value in Python. We can pass either the actual data or a variable name into this function.

For example:

* `type(10)` will return the data type of the value 10
* `type(a)` will return the data type of the value stored in the variable `a`


````python
print(type(10))
print(type(a))
print(type(3.14))
print(type(b))
print(type("R"))
print(type(c))
print(type(True))
print(type(d))
````

# Input Function

The `input()` function is used to take input from the user during the execution of a program.

In Python, the correct and most common way to receive input from the user is by using the `input()` function.

Inside the `input()` function, we can provide a message to the user. This message is optional and is mainly used to instruct or guide the user about what value needs to be entered.

The value received using the `input()` function should always be stored in a variable.

By default, the `input()` function treats every value entered by the user as a **string (`str`)** data type.

For example, even if the user enters:

```python
10
```

Python will internally treat it as:

```python
"10"
```

Therefore, if we need input in a specific data type such as `int` or `float`, we must explicitly convert the input into the required data type while taking the input itself.

---

# Example

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Name:", name)
print("Age:", age)

print(type(name))
print(type(age))
```

### Output

```python
Enter your name: Rahul
Enter your age: 20

Name: Rahul
Age: 20

<class 'str'>
<class 'int'>
```

# Key Points

- `input()` is used to take input from the user
- The message inside `input()` is optional
- Input values should be stored in variables
- By default, every input is treated as a string
- Use:
  - `int()` for integer input
  - `float()` for decimal input
  - `str()` for text input

# Operators in Python

Operators are special symbols used to perform operations on variables and values.

Python provides different types of operators for performing mathematical calculations, comparisons, assignments, and logical operations.

---

# Types of Operators in Python

## 1. Arithmetic Operators

Arithmetic operators are used to perform mathematical operations.

| Operator | Description | Example |
|---|---|---|
| `+` | Addition | `10 + 5` |
| `-` | Subtraction | `10 - 5` |
| `*` | Multiplication | `10 * 5` |
| `/` | Division | `10 / 5` |
| `//` | Floor Division | `10 // 3` |
| `%` | Modulus (Remainder) | `10 % 3` |
| `**` | Exponent (Power) | `2 ** 3` |

---

# Arithmetic Operator Example

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

### Output

```python
Addition: 13
Subtraction: 7
Multiplication: 30
Division: 3.3333333333333335
Floor Division: 3
Modulus: 1
Exponent: 1000
```

---

## 2. Comparison Operators

Comparison operators are used to compare two values.

The result of a comparison operator is always either `True` or `False`.

| Operator | Description | Example |
|---|---|---|
| `==` | Equal to | `a == b` |
| `!=` | Not equal to | `a != b` |
| `>` | Greater than | `a > b` |
| `<` | Less than | `a < b` |
| `>=` | Greater than or equal to | `a >= b` |
| `<=` | Less than or equal to | `a <= b` |

---

# Comparison Operator Example

```python
a = 10
b = 5

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

### Output

```python
False
True
True
False
True
False
```

---

## 3. Logical Operators

Logical operators are used to combine multiple conditions.

| Operator | Description |
|---|---|
| `and` | Returns `True` if both conditions are true |
| `or` | Returns `True` if at least one condition is true |
| `not` | Reverses the result |

---

# Logical Operator Example

```python
a = 10
b = 5

print(a > 5 and b < 10)
print(a < 5 or b < 10)
print(not(a > b))
```

### Output

```python
True
True
False
```

---

## 4. Assignment Operators

Assignment operators are used to assign values to variables.

| Operator | Example | Equivalent To |
|---|---|---|
| `=` | `x = 5` | Assign value |
| `+=` | `x += 2` | `x = x + 2` |
| `-=` | `x -= 2` | `x = x - 2` |
| `*=` | `x *= 2` | `x = x * 2` |
| `/=` | `x /= 2` | `x = x / 2` |

---

# Assignment Operator Example

```python
x = 10

x += 5
print(x)

x *= 2
print(x)
```

### Output

```python

30
```

# Operator Precedence

Operator precedence defines the order in which operations are performed in an expression.

Python follows a specific order while evaluating expressions.

Example:

```python
result = 10 + 5 * 2
print(result)
```

### Output

```python
20
```

Here, multiplication happens before addition.

Using brackets changes the precedence:

```python
result = (10 + 5) * 2
print(result)
```

### Output

```python
30
```

---

# PEMDAS / PUMASREBLTAC Rule

A popular mnemonic used to remember operator precedence is:

## **PUMASREBLTAC**

| Letter | Meaning |
|---|---|
| P | Parentheses `()` |
| U | Unary Operators |
| M | Multiplication `*` |
| A | Addition `+` |
| S | Subtraction `-` |
| R | Relational Operators `<, >, <=, >=` |
| E | Equality Operators `==, !=` |
| B | Bitwise Operators |
| L | Logical Operators |
| T | Ternary Operator |
| A | Assignment Operators |
| C | Comma `,` |

Operators with higher precedence are evaluated first.
