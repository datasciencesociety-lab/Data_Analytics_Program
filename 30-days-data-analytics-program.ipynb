To work with data in any programming language, we need to store that data in the computer’s memory. However, we cannot access memory directly. So, we assign a name to a memory location where the data is stored. This allows us to easily access and use the data whenever needed, as long as it remains in memory.

This named memory location is called a **variable**.

**Definition:**
A variable is a named memory location used to store data.


```python
a = 10
```

`a = 10`

In the above code, **`a`** is the variable name, and **`10`** is the value assigned to it. This means the value 10 is stored in a memory location, and we can access it later using the variable name `a`.


```python
print(a)
```

To display a value on the screen in Python, we use the `print()` function. By passing the variable name inside the `print()` function, we can see the value stored in that variable.

For example, if we write `print(a)`, it will display the value assigned to `a` on the screen.


**Data Types**

As the name suggests, data types tell us the type of data that is being stored in a variable.

In Python, the most commonly used data types are:

* **int** → for integer values (e.g., 10, 25)
* **float** → for decimal values (e.g., 10.5, 3.14)
* **str** → for text or characters (e.g., "hello")

Python does not have a separate **char** data type because a single character is also considered a string.

Similarly, unlike some other programming languages:

* There is no separate **long int** or **short int** — all come under **int**
* There is no separate **double** or **long double** — all come under **float**

This makes Python simpler and easier to use.


```python
a = 10
b = 3.14
c = "Hello World!!"
```

```python
print("Data type of a = ",a) #Int Data type
print("Data type of b = ",b) #Float Data type
print("Data type of c = ",c) #String Data type
```

The `type()` function is used to find the data type of a value in Python. We can pass either the actual data or a variable name into this function.

For example:

* `type(10)` will return the data type of the value 10
* `type(a)` will return the data type of the value stored in the variable `a`


```python
print(type(10))
print(type(a))
print(type(3.14))
print(type(b))
print(type("R"))
print(type(c))
```
