# Why Do We Need a Database?

Imagine you are working for an e-commerce company.

Every day the company stores information about:

- Customers
- Products
- Orders
- Payments
- Reviews

After a few years, the company may have **millions of records**.

Managing such a huge amount of data using Excel or CSV files becomes difficult.

Some common challenges include:

- Finding specific information quickly.
- Storing large amounts of data.
- Avoiding duplicate records.
- Allowing multiple users to access the same data.
- Keeping the data organized.

To solve these problems, companies use **Databases**.

---

# What is a Database?

A **Database** is an organized collection of related data that is stored electronically and can be accessed, managed, and updated efficiently.

Simply put,

> **A database is a place where data is stored in an organized manner.**

---

# What is a Table?

A **Table** stores information about one specific entity.

For example,

- Customers
- Products
- Orders
- Payments

Each table consists of **Rows** and **Columns**.

---

# What is a Row?

A **Row** represents one complete record.

Example:

| CustomerID | Name | City |
|------------|------|------|
| 101 | Rahul | Hyderabad |

The above record represents **one customer**.

---

# What is a Column?

A **Column** represents one attribute of the data.

Example:

| Column | Meaning |
|---------|---------|
| CustomerID | Customer ID |
| Name | Customer Name |
| City | Customer City |

Each column stores one specific type of information.

---

# Database Structure

Imagine an Online Shopping Database.

```
ECommerce Database
│
├── Customers
├── Products
├── Orders
└── Payments
```

Each table stores different information, but together they form one database.

---

# What is DBMS?

**DBMS (Database Management System)** is software used to create, manage, store, and retrieve data from databases.

A DBMS helps us:

- Create databases
- Store data
- Retrieve data
- Update data
- Delete data

Think of it as software that helps us manage databases.

---

# What is RDBMS?

**RDBMS (Relational Database Management System)** is a type of DBMS that stores data in **related tables**.

Instead of storing everything in one large table, an RDBMS organizes data into multiple tables and connects them using relationships.

This makes the data:

- Organized
- Consistent
- Easy to retrieve
- Easy to maintain

Most modern companies use **RDBMS**.

---

# DBMS vs RDBMS

| DBMS | RDBMS |
|------|--------|
| Stores data | Stores data in related tables |
| Relationships are not mandatory | Tables are connected using relationships |
| Suitable for small applications | Suitable for large business applications |
| Less structured | More structured and organized |

---

# Popular Relational Database Systems

Some commonly used RDBMS are:

- MySQL
- PostgreSQL
- SQLite
- Microsoft SQL Server
- Oracle Database

> **In this course, we will use SQLite inside Jupyter Notebook because it is lightweight and easy to learn.**

---

# CRUD Operations

Every application performs four basic operations on data.

These operations are called **CRUD**.

| CRUD | Meaning | SQL Command |
|------|---------|-------------|
| Create | Add new data | INSERT |
| Read | Retrieve existing data | SELECT |
| Update | Modify existing data | UPDATE |
| Delete | Remove existing data | DELETE |

### Example

Imagine an Online Shopping Application.

**Create**

A new customer registers.

➡️ Customer data is added.

**Read**

The customer views their order history.

➡️ Data is retrieved.

**Update**

The customer changes their address.

➡️ Existing data is modified.

**Delete**

The customer deletes their account.

➡️ Data is removed.

> **As Data Analysts, we mainly use the Read operation (`SELECT`) because our job is to analyze existing data.**

---

# Categories of SQL Commands

SQL commands are grouped into different categories.

| Category | Purpose | Commands |
|----------|---------|----------|
| DDL (Data Definition Language) | Defines database structure | CREATE, ALTER, DROP |
| DML (Data Manipulation Language) | Modifies data | INSERT, UPDATE, DELETE |
| DQL (Data Query Language) | Retrieves data | SELECT |

> Don't worry about memorizing these categories now.
>
> We will learn each command practically in the upcoming chapters.

---

# Key Takeaways

- A database stores related information in an organized manner.
- A table stores information about one entity.
- A row represents one record.
- A column represents one attribute.
- DBMS is software used to manage databases.
- RDBMS stores data in related tables.
- CRUD represents the four basic operations performed on data.
- SQL commands are grouped into DDL, DML, and DQL.

# Setting Up SQLite in Jupyter Notebook

SQLite is a lightweight Relational Database Management System (RDBMS).

Unlike MySQL or PostgreSQL, SQLite does not require a separate server installation.

We will use Python's built-in **sqlite3** library to interact with SQLite directly from Jupyter Notebook.

---

## Step 1: Import Required Libraries

```python
import sqlite3
import pandas as pd
```

---

## Step 2: Create a Database

```python
conn = sqlite3.connect("ecommerce.db")
cursor = conn.cursor()
```

If the database does not exist, SQLite automatically creates it.

If it already exists, SQLite simply connects to it.

---

# Creating Our First Table

Before storing data, we need to create a table.

The SQL syntax is:

```sql
CREATE TABLE table_name (
    column_name datatype,
    column_name datatype
);
```

---

# Understanding SQL Data Types

Some commonly used SQL data types are:

| Data Type | Description          |
| --------- | -------------------- |
| INTEGER   | Whole numbers        |
| REAL      | Decimal numbers      |
| TEXT      | Text or Strings      |
| DATE      | Dates                |
| BOOLEAN   | True or False values |

---

# What is a Primary Key?

A **Primary Key** uniquely identifies every row in a table.

A Primary Key:

* Cannot contain duplicate values.
* Cannot contain NULL values.
* Uniquely identifies each record.

Example:

| CustomerID | Name  |
| ---------- | ----- |
| 101        | Rahul |
| 102        | Priya |

Here, **CustomerID** is the Primary Key.

---

# Creating the Customers Table

```python
cursor.execute("""
CREATE TABLE Customers(
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT,
    City TEXT,
    Age INTEGER
);
""")

conn.commit()
```

---

# Creating the Products Table

```python
cursor.execute("""
CREATE TABLE Products(
    ProductID INTEGER PRIMARY KEY,
    ProductName TEXT,
    Category TEXT,
    Price REAL
);
""")

conn.commit()
```

---

# Creating the Orders Table

```python
cursor.execute("""
CREATE TABLE Orders(
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    ProductID INTEGER,
    Quantity INTEGER
);
""")

conn.commit()
```

> **Note:** We will learn how `CustomerID` and `ProductID` connect different tables when we study **Joins**.
