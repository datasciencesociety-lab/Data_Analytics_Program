# What is Inferential Statistics?

Inferential Statistics is the branch of statistics that helps us make conclusions or predictions about an entire population using only a sample of data.

Instead of collecting data from every individual, we analyze a smaller representative sample and use statistical techniques to make informed decisions.

In simple words,

> **Inferential Statistics = Making decisions about a large population using a small sample.**

---

# Why Do We Need Inferential Statistics?

In many real-world situations, collecting data from every individual is:

- Expensive
- Time-consuming
- Sometimes impossible

For example,

Suppose Amazon wants to know the average delivery satisfaction of all its customers.

Would it be practical to ask every customer?

**No.**

Instead, Amazon collects feedback from a sample of customers and uses Inferential Statistics to estimate the satisfaction level of all customers.

This saves:

- Time
- Money
- Resources

while still providing reliable insights.

---

# Descriptive Statistics vs Inferential Statistics

| Descriptive Statistics | Inferential Statistics |
|-------------------------|-------------------------|
| Summarizes existing data | Makes conclusions from data |
| Describes what happened | Predicts or estimates what is likely true |
| Works with the available dataset | Uses a sample to understand a population |
| Uses charts, averages, and summaries | Uses probability and statistical tests |
| Example: Average salary of 500 employees | Example: Estimate average salary of all employees in the company |

---

This decision is made using only a sample instead of surveying everyone.

# Why Do We Need Sampling?

Imagine you are working as a Data Analyst at an e-commerce company.

---

# What is Population?

A **Population** is the complete collection of individuals, objects, or observations that we want to study.

In simple words,

> **Population = The entire group we are interested in.**

### Examples

| Study | Population |
|--------|------------|
| Student Performance | All students in the college |
| Customer Satisfaction | All customers of Amazon |
| Election Survey | All eligible voters |
| Product Quality | Every product manufactured today |

---

# What is a Sample?

A **Sample** is a smaller group selected from the population.

The sample should represent the population as closely as possible.

In simple words,

> **Sample = A small portion of the population.**

### Examples

| Population | Sample |
|------------|--------|
| 10,000 students | 500 students |
| 5 million customers | 2,000 customers |
| All voters in India | 5,000 surveyed voters |
| 50,000 manufactured products | 300 inspected products |

---

# Population vs Sample

| Population | Sample |
|------------|--------|
| Entire group | Small subset |
| Usually very large | Smaller and manageable |
| Expensive to collect | Less expensive |
| Time-consuming | Faster to collect |
| Gives exact values | Used to estimate population values |

---

# Parameter vs Statistic

These two terms are often confused.

## Parameter

A **Parameter** is a numerical value calculated from the **Population**.

Examples:

- Population Mean
- Population Standard Deviation
- Population Proportion

Since collecting data from the entire population is difficult, parameters are often unknown.

---

## Statistic

A **Statistic** is a numerical value calculated from a **Sample**.

Examples:

- Sample Mean
- Sample Standard Deviation
- Sample Proportion

Statistics are used to estimate population parameters.

---

# Parameter vs Statistic

| Parameter | Statistic |
|-----------|-----------|
| Calculated from Population | Calculated from Sample |
| Usually unknown | Can be calculated |
| Represents the whole population | Represents only the sample |

---

# What is Sampling?

Sampling is the process of selecting a subset of individuals from a population for analysis.

The goal of sampling is to make reliable conclusions about the entire population while saving:

- Time
- Cost
- Effort

---

# Sampling Techniques

There are many ways to select a sample from a population.

In this course, we will learn four commonly used sampling techniques. To make them easier to understand, we'll first use simple names and then learn their official statistical names.

---

# 1. Random Sampling (Simple Random Sampling)

Every individual in the population has an equal chance of being selected.

### Example

A college has **5,000 students**.

A computer randomly selects **300 students**.

Every student has an equal probability of being chosen.

### Advantages

* Fair selection
* Reduces selection bias
* Easy to understand and implement

---

# 2. Category Sampling (Stratified Sampling)

The population is first divided into different categories (called **strata**).

A sample is then selected from each category.

### Example

Suppose a college has:

* 60% Boys
* 40% Girls

To maintain the same proportion in the sample, students are selected from both groups.

This ensures every category is represented.

### Advantages

* Better representation of the population
* Useful when the population has different categories

---

# 3. Interval Sampling (Systematic Sampling)

Individuals are selected at regular intervals.

### Example

Suppose there are **10,000 customers**.

Instead of choosing customers randomly, we select:

* Every 50th customer

For example:

50th

100th

150th

200th

...

### Advantages

* Simple and fast
* Easy to implement
* Useful for large datasets

---

# 4. Easy Sampling (Convenience Sampling)

Individuals are selected based on their availability or ease of access.

### Example

A teacher wants quick feedback and asks only the students sitting in the first two rows.

### Advantages

* Very fast
* Easy to collect data
* Requires minimal effort

### Disadvantage

The sample may not represent the entire population, leading to biased results.

---

# Comparison of Sampling Techniques

| Easy Name         | Official Name          | Selection Method                    | Best Used When                  |
| ----------------- | ---------------------- | ----------------------------------- | ------------------------------- |
| Random Sampling   | Simple Random Sampling | Completely random                   | General surveys                 |
| Category Sampling | Stratified Sampling    | Select from every category          | Population has different groups |
| Interval Sampling | Systematic Sampling    | Select at fixed intervals           | Large ordered datasets          |
| Easy Sampling     | Convenience Sampling   | Select easily available individuals | Quick surveys                   |

---

# Key Takeaways

* Sampling is the process of selecting a subset from a population.
* Different sampling techniques are used for different situations.
* Random Sampling gives everyone an equal chance.
* Category Sampling ensures every group is represented.
* Interval Sampling selects individuals at fixed intervals.
* Easy Sampling is quick but may introduce bias.

# What is Probability?

Probability is the branch of mathematics that measures the likelihood of an event occurring.

In simple words,

> **Probability tells us how likely something is to happen.**

For example:

- What is the probability that it rains tomorrow?
- What is the probability that a customer buys a product?
- What is the probability of getting a Head when tossing a coin?

Probability helps us measure uncertainty.

---

# Probability Range

The value of probability always lies between **0 and 1**.

```text
0 ≤ P(Event) ≤ 1
```

Where:

- **0** → Impossible Event
- **1** → Certain Event

---

# Why is Probability Important?

As Data Analysts, we often answer questions like:

- What is the chance that a customer will purchase a product?
- What is the chance that an order will be delayed?
- What is the chance that a marketing campaign will succeed?
- What is the chance that our sample result happened just by random chance?

Probability helps us answer these questions and forms the foundation of Inferential Statistics.

---

# Basic Terminology

Before learning probability calculations, let's understand some important terms.

---

# Experiment

An **Experiment** is an activity or process that produces one or more outcomes.

### Examples

- Tossing a coin
- Rolling a dice
- Drawing a card
- Selecting a customer

---

# Outcome

An **Outcome** is a single possible result of an experiment.

Example

Experiment:

```text
Rolling a Dice
```

Possible Outcomes

```text
1,2,3,4,5,6
```

Each number is called an **Outcome**.

---

# Event

An **Event** is one or more outcomes that satisfy a condition.

Example

Experiment:

```text
Rolling a Dice
```

Event:

```text
Getting an Even Number
```

Possible Outcomes

```text
2,4,6
```

---

# Sample Space

The **Sample Space** is the collection of all possible outcomes.

It is represented by:

```text
S
```

Example

Coin Toss

```text
S = {Head, Tail}
```

Rolling a Dice

```text
S = {1,2,3,4,5,6}
```

---

# Probability Formula

```text
            Number of Favorable Outcomes
P(Event) = ------------------------------
             Total Number of Outcomes
```

---

# Example 1

A coin is tossed.

Find the probability of getting a Head.

Sample Space

```text
{Head, Tail}
```

Favorable Outcomes

```text
{Head}
```

Therefore,

```text
P(Head)=1/2=0.5
```

---

# Example 2

A fair dice is rolled.

Find the probability of getting 5.

Sample Space

```text
{1,2,3,4,5,6}
```

Favorable Outcomes

```text
{5}
```

Therefore,

```text
P(5)=1/6
```

---

# Example 3 (Business Example)

A company sends promotional emails to **1000 customers**.

250 customers click the email.

Find the probability that a randomly selected customer clicked the email.

Solution

```text
P(Clicked)

=250/1000

=0.25
```

---

# Addition Rule of Probability

The Addition Rule is used when finding the probability of:

> **Event A OR Event B**

Example

Rolling a dice.

Find the probability of getting:

```text
2 OR 5
```

Solution

```text
P(2 OR 5)

=2/6

=1/3
```

---

# Multiplication Rule of Probability

The Multiplication Rule is used when finding the probability of:

> **Event A AND Event B**

Example

Coin Toss + Dice Roll

Find the probability of getting:

```text
Head AND 6
```

Solution

```text
P(Head AND 6)

=(1/2)×(1/6)

=1/12
```

---

# Independent Events

Two events are called Independent if the occurrence of one event does not affect the occurrence of the other.

Examples

- Tossing a coin and rolling a dice
- Weather today and rolling a dice

---

# Why Do We Learn Probability?

Probability is one of the building blocks of Inferential Statistics.

It helps us:

- Measure uncertainty
- Estimate unknown values
- Perform Hypothesis Testing
- Calculate P-values
- Build Confidence Intervals

Without Probability, Inferential Statistics would not be possible.

---

# Key Takeaways

- Probability measures uncertainty.
- Probability values always lie between 0 and 1.
- Experiments produce Outcomes.
- Events are collections of Outcomes.
- Sample Space contains all possible Outcomes.
- Addition Rule is used for OR events.
- Multiplication Rule is used for independent AND events.
- Probability is the foundation of Inferential Statistics.

# What is a Distribution?

A **Distribution** describes how data values are spread.

It tells us:

- Where most values are located.
- Whether values are concentrated or spread out.
- Whether there are extreme values (outliers).

In simple words,

> **A distribution shows how data is distributed across different values.**

---

# Example

Suppose we record the marks of 100 students.

Some students score:

- Very high marks
- Most students score around the average
- Some students score very low

A distribution helps us visualize this pattern.

---

# What is a Normal Distribution?

A **Normal Distribution**, also called the **Gaussian Distribution**, is one of the most common probability distributions.

It forms a **bell-shaped curve**, where:

- Most observations lie near the center.
- Fewer observations lie far from the center.
- The left and right sides are symmetrical.

It is represented as a smooth, bell-shaped curve.

---

# Characteristics of a Normal Distribution

A Normal Distribution has the following properties:

- Bell-shaped curve
- Symmetrical about the center
- Mean = Median = Mode
- Most observations lie near the center
- Fewer observations occur as we move away from the center
- The total area under the curve equals 1 (or 100%)

---

# Bell Curve

A Normal Distribution is often called a **Bell Curve** because of its shape.

```text
                *
             *     *
           *         *
         *             *
       *                 *
     *                     *
   *                         *
 *                             *
-------------------------------------------->
```

The highest point of the curve represents the value where most observations occur.

---

# Mean, Median and Mode

In a perfectly Normal Distribution,

```text
Mean = Median = Mode
```

All three measures are located exactly at the center of the distribution.

---

# The Empirical Rule (68–95–99.7 Rule)

The Empirical Rule describes how data is distributed in a Normal Distribution.

### About 68% of data

lies within **1 Standard Deviation** from the Mean.

---

### About 95% of data

lies within **2 Standard Deviations** from the Mean.

---

### About 99.7% of data

lies within **3 Standard Deviations** from the Mean.

---

## Summary

| Range | Percentage of Data |
|--------|--------------------:|
| Mean ± 1 Standard Deviation | 68% |
| Mean ± 2 Standard Deviations | 95% |
| Mean ± 3 Standard Deviations | 99.7% |

---

# Real-Life Examples of Normal Distribution

Many real-world datasets approximately follow a Normal Distribution.

Examples include:

- Students' heights
- IQ scores
- Blood pressure
- Birth weights
- Measurement errors
- Exam scores (in many large exams)

Not every dataset follows a Normal Distribution, but many statistical methods assume that the data is approximately normal.

---

# Why is Normal Distribution Important?

Many statistical techniques are built on the assumption that data follows a Normal Distribution.

Examples include:

- Confidence Intervals
- Hypothesis Testing
- Z-Test
- T-Test

Understanding the Normal Distribution helps us make reliable statistical decisions.

---

# Business Example

Suppose a company tracks the daily delivery time of thousands of orders.

Most deliveries take around **3 days**.

Some deliveries are slightly faster.

Some deliveries are slightly slower.

Very few deliveries take an unusually long time.

If the delivery times follow a Normal Distribution, analysts can estimate:

- Typical delivery time
- Unusual delays
- Expected customer experience

---

# Key Takeaways

- A Distribution shows how data is spread.
- A Normal Distribution is a bell-shaped, symmetrical distribution.
- Mean, Median and Mode are equal in a Normal Distribution.
- Most observations lie near the center.
- The Empirical Rule explains how data is distributed.
- Many inferential statistical methods assume normally distributed data.

# Why Can't We Trust Every Sample?

Every sample is different.

For example,

Sample 2:

Average Rating = **4.5**

Sample 3:

Average Rating = **4.4**

Sample 4:

Average Rating = **4.3**

Different samples produce different results.

Inferential Statistics helps us decide whether these differences are meaningful or simply due to random variation.

# What is a Hypothesis?

A **Hypothesis** is an assumption or claim that can be tested using data.

In simple words,

> **A hypothesis is an educated assumption that we want to verify using evidence.**

Notice that a hypothesis is **not automatically true or false**.

It is simply a statement that requires testing.

---

# Everyday Examples of a Hypothesis

Before learning business examples, let's look at some everyday situations.

### Example 1

"I think students who study every day score better marks."

This is a hypothesis.

---

### Example 2

"I believe drinking more water improves concentration."

This is also a hypothesis.

---

### Example 3

"I think online classes improve learning."

Again,

this is a hypothesis.

None of these statements are proven yet.

They require data.

---

# What is Hypothesis Testing?

Hypothesis Testing is the process of checking whether a hypothesis is supported by data.

Instead of making decisions based on opinions, we use statistical evidence.

In simple words,

> **Hypothesis Testing is a statistical method used to determine whether a business claim is supported by sample data.**

---

# Why Do Businesses Use Hypothesis Testing?

Businesses make assumptions every day.

Examples include:

* Our new website increased customer satisfaction.
* The new marketing campaign improved sales.
* Average delivery time is below 5 days.
* Customers prefer Product A over Product B.

These are only claims.

Before taking important business decisions, companies need evidence.

Hypothesis Testing provides that evidence.

---

# Business Example

Suppose an online shopping company launches a new mobile application.

After one month, customer ratings appear higher.

Management asks:

> "Has the new app actually improved customer satisfaction?"

Instead of guessing,

we perform Hypothesis Testing.

---

# Two Possible Situations

Whenever we test a business claim, there are only two possibilities.

### Situation 1

The business claim is **not true**.

The observed difference happened only because of random chance.

---

### Situation 2

The business claim is **supported by the data**.

The observed difference is statistically significant.

Hypothesis Testing helps us decide which situation is more likely.

---

# Null Hypothesis (H₀)

The **Null Hypothesis** represents the default assumption.

It usually states that:

* Nothing has changed.
* There is no difference.
* There is no effect.

Think of it as the position we assume to be true until sufficient evidence suggests otherwise.

### Examples

* Average customer rating is **4.0**.
* The new website has **no effect** on customer satisfaction.
* Average delivery time is **5 days**.

---

# Alternative Hypothesis (H₁)

The **Alternative Hypothesis** represents the business claim that we want to investigate.

It usually states that:

* Something has changed.
* A difference exists.
* An effect is present.

### Examples

* Average customer rating is **greater than 4.0**.
* The new website improves customer satisfaction.
* Average delivery time is **less than 5 days**.

---

# Understanding H₀ and H₁ Together

Suppose a company claims:

> "The average customer rating is greater than 4."

We write:

```text
H₀ : Average Rating = 4

H₁ : Average Rating > 4
```

We collect sample data and perform a statistical test.

The result helps us decide whether there is enough evidence to reject the Null Hypothesis.

---

# Another Example

Business Question:

> "Has the new marketing campaign increased sales?"

```text
H₀ : The campaign has no effect on sales.

H₁ : The campaign increases sales.
```

Again,

we collect data before making any decision.

---

# Steps in Hypothesis Testing

Every Hypothesis Test follows the same general process.

```text
Business Problem
        ↓
Business Claim
        ↓
Create H₀ and H₁
        ↓
Collect Sample Data
        ↓
Perform Statistical Test
        ↓
Calculate P-value
        ↓
Make Decision
        ↓
Business Recommendation
```

This is the complete Hypothesis Testing workflow.

---

# Important Note

Hypothesis Testing does **not** prove that a hypothesis is true.

Instead,

it tells us whether there is **enough statistical evidence** to support the business claim.

This is an important distinction.

---

# Key Takeaways

* A Hypothesis is a testable assumption.
* Hypothesis Testing uses sample data to evaluate a business claim.
* Every test begins with two hypotheses:

  * Null Hypothesis (H₀)
  * Alternative Hypothesis (H₁)
* The Null Hypothesis represents the default assumption.
* The Alternative Hypothesis represents the claim being investigated.
* Hypothesis Testing helps businesses make evidence-based decisions.

# Chapter 3: P-value and Significance Level

---

# Learning Objectives

By the end of this chapter, you will be able to:

* Understand why Data Analysts need statistical evidence before making business decisions.
* Learn what a Statistical Test produces.
* Understand the concept of a P-value.
* Understand the Significance Level (α).
* Learn how statistical decisions are made.
* Interpret statistical results from a business perspective.

---

# From Business Question to Business Decision

In the previous chapter, we learned how to formulate:

* Null Hypothesis (H₀)
* Alternative Hypothesis (H₁)

These hypotheses represent two possible explanations for a business problem.

However, writing H₀ and H₁ does not answer the business question.

As Data Analysts, our job is to determine **which hypothesis is better supported by the available evidence.**

To do this, we perform a **Statistical Test**.

The Statistical Test analyzes the sample data and helps us decide whether the observed result is statistically convincing.

---

# The Role of Statistical Testing

Consider the following business claim:

> "The new delivery process has reduced the average delivery time below 5 days."

Suppose we collect delivery information from a sample of customers.

After analyzing the sample, we obtain an average delivery time of **4.8 days**.

At first glance, this seems to support the business claim.

But an important question remains:

> **Is this difference actually meaningful, or could it simply be the result of random sampling?**

Since our conclusion is based on a sample rather than the entire population, we need a method to evaluate the strength of the evidence.

That method is **Statistical Testing**.

---

# What Does a Statistical Test Produce?

A Statistical Test analyzes the sample data and produces an important value called the **P-value**.

The P-value is not the final decision.

Instead, it helps us measure how strongly the sample data disagrees with the Null Hypothesis.

The stronger the evidence against the Null Hypothesis, the more confident we become that the observed result is unlikely to have occurred by chance.

---

# Understanding the P-value

The **P-value** is a numerical measure that tells us how consistent our sample data is with the Null Hypothesis.

In simple terms,

> **The P-value indicates the strength of evidence against the Null Hypothesis.**

Think of the P-value as an **evidence score**.

* A **small P-value** indicates strong evidence against H₀.
* A **large P-value** indicates weak evidence against H₀.

The P-value does **not** tell us whether a hypothesis is true or false.

Instead, it helps us decide whether the observed sample evidence is strong enough to reject the Null Hypothesis.

---

# Why Do We Need a Decision Rule?

Imagine three Data Analysts analyzing the same business problem.

The first analyst rejects the Null Hypothesis when the P-value is **0.08**.

The second rejects it only when the P-value is **0.04**.

The third rejects it only when the P-value is **0.01**.

If every analyst follows a different rule, the organization may reach different conclusions from the same data.

To ensure consistency, organizations use a predefined decision threshold called the **Significance Level**.

---

# What is the Significance Level (α)?

The **Significance Level**, represented by the symbol **α (alpha)**, is the threshold used to determine whether the available evidence is strong enough to reject the Null Hypothesis.

In most business and research applications,

```text
α = 0.05
```

This value means that we require a reasonable level of statistical evidence before concluding that the observed result is significant.

For this course, we will use:

```text
α = 0.05
```

---

# Making a Statistical Decision

Once the Statistical Test produces a P-value, we compare it with the Significance Level.

### Case 1

If

```text
P-value < α
```

the evidence provided by the sample is considered strong enough.

Decision:

```text
Reject H₀
```

Business Interpretation:

The sample provides sufficient statistical evidence to support the business claim represented by the Alternative Hypothesis.

---

### Case 2

If

```text
P-value ≥ α
```

the evidence provided by the sample is not strong enough.

Decision:

```text
Fail to Reject H₀
```

Business Interpretation:

The available sample does not provide sufficient evidence to support the business claim.

---

# Example 1

### Business Problem

An e-commerce company claims that its new recommendation system has increased the average customer rating above **4.0**.

After collecting sample data, a Statistical Test returns:

```text
P-value = 0.02
```

The company uses:

```text
α = 0.05
```

Since

```text
0.02 < 0.05
```

Decision:

```text
Reject H₀
```

### Business Interpretation

The sample provides strong statistical evidence that the average customer rating is greater than 4.0.

Based on this analysis, the company has evidence supporting the effectiveness of the new recommendation system.

---

# Example 2

### Business Problem

A logistics company believes that its new delivery process has reduced the average delivery time below **5 days**.

The Statistical Test returns:

```text
P-value = 0.31
```

Using

```text
α = 0.05
```

Since

```text
0.31 ≥ 0.05
```

Decision:

```text
Fail to Reject H₀
```

### Business Interpretation

The available sample does not provide sufficient statistical evidence to conclude that the average delivery time has decreased below 5 days.

This does **not** prove that the new process failed.

It simply means that the current sample does not provide enough evidence to support the claim.

---

# Important Points to Remember

Statistical Testing helps us make decisions using evidence rather than assumptions.

However, there are some important limitations.

A Statistical Test does **not**:

* Prove that the Null Hypothesis is true.
* Prove that the Alternative Hypothesis is absolutely true.
* Guarantee that the business decision is correct.

Instead, it measures how strongly the sample data supports or contradicts the Null Hypothesis.

Therefore, our conclusions are always written as:

* Reject H₀

or

* Fail to Reject H₀

We never say:

* Accept H₀

because our conclusions are based only on sample data.

---

# Summary

The complete decision-making process can be summarized as:

```text
Business Question
        ↓
Collect Sample Data
        ↓
Perform Statistical Test
        ↓
Obtain P-value
        ↓
Compare with α
        ↓
Statistical Decision
        ↓
Business Interpretation
        ↓
Business Recommendation
```

As Data Analysts, our objective is not simply to calculate a P-value.

Our objective is to use statistical evidence to support better business decisions.

---

# Key Takeaways

* Statistical Tests produce a P-value.
* The P-value measures the strength of evidence against the Null Hypothesis.
* The Significance Level (α) is the decision threshold.
* If P-value < α → Reject H₀.
* If P-value ≥ α → Fail to Reject H₀.
* Every statistical decision should ultimately be translated into a meaningful business recommendation.

---

# Can Our Decision Be Wrong?

In the previous chapter, we learned how to make a statistical decision using the P-value.

But remember,

we are making decisions based on a **sample**, not the entire population.

Because of this,

there is always a possibility of making the wrong decision.

This is completely normal in statistics.

---

# Why Do Errors Happen?

Suppose we survey only **500 customers** out of **5 million customers**.

The sample may not perfectly represent the entire population.

As a result,

our statistical decision might sometimes be incorrect.

These incorrect decisions are called **Statistical Errors**.

---

# Two Types of Errors

Whenever we perform Hypothesis Testing, there are two possible mistakes.

1. Type I Error
2. Type II Error

Let's understand both one by one.

---

# Type I Error (False Positive)

A **Type I Error** happens when we **Reject the Null Hypothesis even though it is actually true.**

In simple words,

> **We conclude that a change exists when it actually does not.**

This is also called a **False Positive**.

---

# Real-Life Example

Suppose a COVID test reports that a healthy person is infected.

Reality:

```text
Healthy
```

Test Result:

```text
Positive
```

This is a **False Positive**.

---

# Business Example

A company believes:

> "The new marketing campaign increased sales."

After performing Hypothesis Testing,

we reject H₀.

The company spends a large marketing budget.

Later,

it turns out the campaign never actually improved sales.

This is a **Type I Error**.

---

# Type II Error (False Negative)

A **Type II Error** happens when we **Fail to Reject the Null Hypothesis even though it is actually false.**

In simple words,

> **We fail to detect a real change.**

This is called a **False Negative**.

---

# Real-Life Example

Suppose a COVID test reports that an infected person is healthy.

Reality:

```text
Infected
```

Test Result:

```text
Negative
```

This is a **False Negative**.

---

# Business Example

A company launches a new recommendation system.

The system actually improves customer satisfaction.

However,

our statistical test concludes that there is no improvement.

The company removes the new system.

A good business decision is lost because of a **Type II Error**.

---

# Comparison

| Type I Error                         | Type II Error                               |
| ------------------------------------ | ------------------------------------------- |
| Reject H₀ when H₀ is actually true   | Fail to Reject H₀ when H₀ is actually false |
| False Positive                       | False Negative                              |
| Detects a change that does not exist | Misses a change that actually exists        |

---

# Easy Way to Remember

### Type I Error

> **False Alarm**

Example:

The fire alarm rings even though there is no fire.

---

### Type II Error

> **Missed Alarm**

Example:

There is a fire, but the alarm never rings.

---

# Why Do Businesses Care?

Making incorrect decisions can be expensive.

Examples:

* Launching an ineffective product.
* Rejecting a useful marketing strategy.
* Approving a poor medical treatment.
* Ignoring a genuine quality issue.

Hypothesis Testing helps reduce these risks, but it cannot eliminate them completely.

---

# Key Takeaways

* Statistical decisions are made using samples, so mistakes are possible.
* Type I Error means rejecting a true Null Hypothesis.
* Type II Error means failing to reject a false Null Hypothesis.
* Type I Error is called a False Positive.
* Type II Error is called a False Negative.
* Businesses aim to reduce both types of errors when making decisions.

# Introduction

So far in this module, we have learned how to:

* Define a business problem.
* Formulate the Null Hypothesis (H₀) and Alternative Hypothesis (H₁).
* Understand uncertainty while making decisions from sample data.

The next step is to determine whether the available sample provides enough evidence to support the business claim.

This is where **Statistical Testing** comes into the picture.

A **Statistical Test** is a statistical procedure that analyzes sample data and helps us decide whether there is enough evidence to reject the Null Hypothesis.

Simply put,

> **A Statistical Test helps us determine whether the observed result is statistically significant or whether it could have occurred simply due to random sampling.**

---

# Why Do We Have Different Statistical Tests?

Not every business problem is the same.

Sometimes we want to answer questions like:

* Is the average customer rating greater than 4?
* Has the average delivery time decreased below 5 days?
* Is there a relationship between customer segment and product category?
* Are customer preferences different across cities?

Each of these questions involves different types of data and different business objectives.

Therefore, **one Statistical Test cannot solve every problem.**

Just as we choose different charts for different visualizations, we choose different Statistical Tests for different analytical problems.

---

# Common Components of Every Statistical Test

Although there are many Statistical Tests, they all follow the same structure.

Every Statistical Test consists of:

* A Business Problem
* Null Hypothesis (H₀)
* Alternative Hypothesis (H₁)
* Sample Data
* A Statistical Test
* Test Statistic
* P-value
* Statistical Decision
* Business Interpretation

The overall process remains the same.

Only the Statistical Test changes depending on the business problem.

---

# General Workflow

Every Statistical Test follows the same workflow.

```text
Business Problem
        ↓
State H₀ and H₁
        ↓
Collect Sample Data
        ↓
Choose the Appropriate Statistical Test
        ↓
Perform the Statistical Test
        ↓
Obtain Test Statistic & P-value
        ↓
Make Statistical Decision
        ↓
Business Interpretation
```

````python
import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(42)

# Number of Records
n_records = 750

# ==============================================================================
# SECTION 1: Create Base Dataset
# ==============================================================================

df = pd.DataFrame({
    'order_id': range(10001, 10001 + n_records),
    'customer_id': np.random.randint(1001, 1501, n_records),
    'customer_segment': np.random.choice(
        ['Premium', 'Standard', 'Budget'],
        n_records,
        p=[0.25, 0.45, 0.30]
    ),
    'product_category': np.random.choice(
        ['Electronics', 'Clothing', 'Home', 'Books', 'Sports'],
        n_records
    )
})

# ==============================================================================
# SECTION 2: Product Price Based on Category
# ==============================================================================

price_ranges = {
    'Electronics': (15000, 80000),
    'Clothing': (500, 5000),
    'Home': (2000, 25000),
    'Books': (200, 2000),
    'Sports': (1000, 15000)
}

df['product_price'] = df['product_category'].apply(
    lambda x: np.random.uniform(
        price_ranges[x][0],
        price_ranges[x][1]
    )
).round(2)

# ==============================================================================
# SECTION 3: Discount Based on Customer Segment
# ==============================================================================

discount_ranges = {
    'Premium': (5, 20),
    'Standard': (0, 10),
    'Budget': (10, 30)
}

df['discount_percent'] = df['customer_segment'].apply(
    lambda x: np.random.uniform(
        discount_ranges[x][0],
        discount_ranges[x][1]
    )
).round(2)

# ==============================================================================
# SECTION 4: Quantity Depends on Discount
# ==============================================================================

df['quantity_ordered'] = df['discount_percent'].apply(
    lambda x: max(
        1,
        int(np.random.normal(3 + (x / 20), 1.5))
    )
)

# ==============================================================================
# SECTION 5: Revenue Calculation
# ==============================================================================

df['total_revenue'] = (
    df['product_price']
    * df['quantity_ordered']
    * (1 - df['discount_percent'] / 100)
).round(2)

# ==============================================================================
# SECTION 6: Delivery Time
# ==============================================================================

df['delivery_days'] = df['product_category'].apply(
    lambda x:
    np.random.randint(8, 14)
    if x == 'Electronics'
    else np.random.randint(2, 7)
)

# ==============================================================================
# SECTION 7: Customer Rating
# ==============================================================================

base_rating = 4.5

df['customer_rating'] = (
    base_rating
    - (df['delivery_days'] - 2) * 0.15
    - (df['discount_percent'] / 100) * 0.3
)

df['customer_rating'] = (
    df['customer_rating']
    + np.random.normal(0, 0.25, n_records)
).clip(1, 5).round(2)

# ==============================================================================
# SECTION 8: City
# ==============================================================================

cities = [
    'Mumbai',
    'Delhi',
    'Bangalore',
    'Hyderabad',
    'Chennai',
    'Kolkata',
    'Pune',
    'Ahmedabad'
]

df['city'] = np.random.choice(
    cities,
    n_records
)

# ==============================================================================
# SECTION 9: Column Order
# ==============================================================================

df = df[[
    'order_id',
    'customer_id',
    'customer_segment',
    'product_category',
    'product_price',
    'quantity_ordered',
    'discount_percent',
    'total_revenue',
    'delivery_days',
    'customer_rating',
    'city'
]]

# ==============================================================================
# SECTION 10: Save Dataset
# ==============================================================================

df.to_csv(
    "ecommerce_dataset.csv",
    index=False
)

# Preview

df.head()
````

---

# What is a One-Sample Z-Test?

A **One-Sample Z-Test** is a statistical test used to determine whether the **mean of a sample** is significantly different from a known or claimed **population mean**.

In simple words,

It answers questions like:

- Is the average delivery time really less than 5 days?
- Is the average customer rating greater than 4?
- Is the average discount equal to 20%?

Instead of guessing, the Z-Test provides statistical evidence.

# When Should We Use a One-Sample Z-Test?

Use a One-Sample Z-Test when:

- We have **one numerical sample**.
- We want to compare the sample mean with a known or claimed population mean.
- The population standard deviation (σ) is known.
- The sample size is sufficiently large (typically n ≥ 30).

---

# When Should We NOT Use a Z-Test?

Do NOT use a Z-Test when:

- The data is categorical.
- The population standard deviation is unknown.
- The sample size is very small.

In these situations, other statistical tests such as the One-Sample T-Test or Chi-Square Test are more appropriate.

---

# Hypothesis Formulation

Business Claim

Average delivery time is less than 5 days.

Null Hypothesis (H₀)

```
μ ≥ 5
```

Alternative Hypothesis (H₁)

```
μ < 5
```

Since the claim is **less than**, this is a **Left-Tailed Test**.

---

# Formula

The Z statistic is calculated as:

\[
Z=\frac{\bar{x}-\mu}{\sigma/\sqrt{n}}
\]

Where,

| Symbol | Meaning |
|---------|----------|
| x̄ | Sample Mean |
| μ | Population Mean |
| σ | Population Standard Deviation |
| n | Sample Size |

---

# Formula Intuition

The formula simply measures:

> **How far is the sample mean from the claimed population mean?**

If the sample mean is very different from the claimed value, the Z-score becomes larger (in magnitude), indicating stronger evidence against the null hypothesis.

Fortunately, Python calculates this automatically.

---

# Python Implementation

Import the required library.

```python
from statsmodels.stats.weightstats import ztest
```
